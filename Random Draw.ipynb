```python
import random
import string

# To fill in
n_jury = 9
n_candidates = 97
number_jury_per_candidate = 5

############ DO NOT TOUCH ###############
jury = list(string.ascii_uppercase)[:n_jury]
candidates = range(1, n_candidates + 1)

candidate_to_jury = {}
for c in candidates:
    candidate_to_jury[c] = random.sample(jury, number_jury_per_candidate)
    
jury_to_candidate = dict([(j, []) for j in jury])

for j in jury:
    for c in candidates:
        if j in candidate_to_jury[c]:
            jury_to_candidate[j].append(c)

for item in jury_to_candidate.items():
    print(item)
```
